# **6. Data Pre-Processing**
---

- This section is emphasised on performing data manipulation over unstructured data for further processing and analysis.

- To modify unstructured data to strucuted data you need to verify and manipulate the integrity of the data by:
  - Handling missing data,

  - Handling redundant data,

  - Handling inconsistent data,

  - Handling outliers,

  - Handling typos

```python
#1)gender has missing value 
```

```python
data["gender"].value_counts()
```

```python
#Identifing blank values 
```

```python
data.shape[0]-data["gender"].count()
```

```python
#filling missing value with mode
data["gender"]=data["gender"].fillna(data["gender"].mode()[0])
```

```python
data["gender"].value_counts()
```

```python
data.shape[0]-data["gender"].count()  #checking blank value
```

```python
# WOKING ON tenure
#FOUND 2 MISSINNG VALUE WE WILL FILL WITH MEAN
# tenure is +ve Sqv..ed
```

```python
data.shape[0]-data["tenure"].count()  #identifing blank values
```

```python
data["tenure"]=data["tenure"].fillna(data["tenure"].mean())  #filling blank values with Mean
```

```python
data.shape[0]-data["tenure"].count()
```

```python
#Working on Age
```

```python
data.shape[0]-data["age"].count()  #checking blank value
```

```python
#Age has no blank values
#Age has oulier 
# 75%+(1.5IQR)=50+45=95
# 25%=1.5IQR)==20+45 ON LEFT SIDE
```

```python
data["age"].nunique()
```

```python
data["age"].unique()
```

```python
data["age"].value_counts().sort_index(ascending=False)
```

```python
# dropping all age above 95
```

```python
data_copy=data#Creating a copy
```

```python
data_copy[(data_copy["age"]>=12)&(data_copy["age"]<=95)]        #Filtering age
```

```python
fb_data=data_copy[(data_copy["age"]>=12)&(data_copy["age"]<=95)]
```

# *8.1  Ploting Male Users Vs Female users

```python
figure = plt.figure(figsize=[15, 7]) 

fb_data['gender'].value_counts().plot.bar(color='#800020') 
plt.xticks(rotation=0, size=12)

plt.yticks(ticks=np.arange(0, 70000, 10000), size=12) 
plt.xlabel(xlabel='Gender', size=14)
plt.ylabel(ylabel='Count', size=14)
plt.title(label='Number of Users', size=16)
plt.show()
```

<a name = Section8></a>
# *8.2 pie plot % of users per Category


```python
figure = plt.figure(figsize=[10, 10]) 
fb_data['gender'].value_counts().plot.pie(autopct='%3.1f%%',startangle=100,legend=True) # extracting the date the ploting pie graph
#% value autopct='%3.1f%%' shows the % value in pie chart


plt.title(label='% of users per Gender', size=16,color='Black')#Labelling title with a custom label and size of 16 with colour

plt.show() # Dispaly the output by rendering visual on the screen
```

<a name = Section8></a>
# *8.3  USER BASED ON AGE


```python
figure=plt.figure(figsize=[15,7])
fb_data['age'].plot.hist(bins=np.arange(10, 90, 5),color="#99004C",grid=True)

plt.xlabel("AGE",size=15)
plt.ylabel("NO OF USERS",size=15)
plt.title("Distribution of USER BASED ON AGE",size=17)

plt.xticks(ticks=np.arange(0, 90, 5), size=12)
plt.yticks(ticks=np.arange(0, 25000, 5000), size=12)
plt.show()
```

# *8.6 Tenure vs Age

```python
figure=plt.figure(figsize=[15,7])

sns.lineplot(data=fb_data, x="age", y="tenure",color='green')


plt.xticks(np.arange(10,120,10),size=12)
plt.yticks(np.arange(0,2100,300),size=12)


plt.xlabel(xlabel='Age', size=14)
plt.ylabel(ylabel='Days', size=14)
plt.title(label='Tenure vs Age', size=16)
plt.show() 
```

<a name = Section8></a>
# *8.7 Tenure vs Age & Gender 

```python
figure=plt.figure(figsize=[15,7])

maleuser=fb_data[fb_data["gender"]=="male"]
femaleuser=fb_data[fb_data["gender"]=="female"]
sns.lineplot(data=maleuser, x="age", y="tenure",color='green')
sns.lineplot(data=femaleuser, x="age", y="tenure",color='red')
plt.legend(labels=('male', 'female'), fontsize=14)
plt.xticks(np.arange(10,120,10),size=12)
plt.yticks(np.arange(0,2100,300),size=12)


plt.xlabel(xlabel='Age', size=14)
plt.ylabel(ylabel='Days Used', size=14)
plt.title(label='Tenure vs Age & Gender', size=16)
plt.show() 
```

```python
#Likes mobile vs web
```

```python
figure = plt.figure(figsize=[15, 7])
fb_data.groupby(by=['gender'])['mobile_likes'].mean().plot.bar(color='grey')
fb_data.groupby(by=['gender'])['www_likes'].mean().plot.bar(color='orange')
plt.xticks(size=12, rotation=90)
plt.yticks(ticks=np.arange(0, 220, 20), size=12)
plt.xlabel(xlabel='Gender', size=14)
plt.ylabel(ylabel='likes', size=14)
plt.title(label='Likes mobile vs web', size=16)
plt.legend(labels=('Web', 'mobile'))
plt.grid(b=True)
plt.show()
```

```python
#Received Likes mobile vs web
```

```python
figure = plt.figure(figsize=[15, 7])
fb_data.groupby(by=['gender'])['mobile_likes_received'].mean().plot.bar(color='grey')
fb_data.groupby(by=['gender'])['www_likes_received'].mean().plot.bar(color='orange')
plt.xticks(size=12, rotation=90)
plt.yticks(ticks=np.arange(0, 250, 50), size=12)
plt.xlabel(xlabel='Gender', size=14)
plt.ylabel(ylabel='likes recieved', size=14)
plt.title(label='Received Likes mobile vs web', size=16)
plt.legend(labels=('Web', 'mobile'))
plt.grid(b=True)
plt.show()
```
